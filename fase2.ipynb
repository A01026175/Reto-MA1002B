Pensamos que los océanos sean infinitos y que los peces sean infinitos, pero en realidad no son inagotables (https://theendofthelinemovie.com/ Links to an external site.).

Existen diferentes modelos de crecimiento de una población (https://es.khanacademy.org/science/ap-biology/ecology-ap/population-ecology-ap/a/exponential-logistic-growth Links to an external site.) que, juntos a otras ecuaciones, como las de Lotka-Volterra, se pueden usar para modelar un sistema biológico.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.01
K = 500
a=1
b=0.02
c=1 
d=0.01

x0=20
y0=20
t0=0 
tf=30 

nt = int((tf-t0)/h) #número de pasos
T=np.arange(0,nt+1)

x = np.empty(nt+1)
x[0]=x0

y = np.empty(nt+1)  
y[0]=y0


for step in range(nt):
    xp = a*x[step]*(1-x[step]/K) - b*x[step]*y[step]
    yp = -c*y[step]+d*x[step]*y[step]

    x[step+1]=x[step]+xp*d
    y[step+1]=y[step]+yp*d
```

```python
plt.plot(x,y)
plt.title("Gráfico de órbitas:")
```

```python
plt.plot(T, x)
plt.plot(T,y)
plt.title("Gráfico de Depredador/Presa")
```

Método ODEINT Recuperado de https://scientific-python.readthedocs.io/en/latest/notebooks_rst/3_Ordinary_Differential_Equations/02_Examples/Lotka_Volterra_model.html

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate

alpha = 1. #mortality rate due to predators
beta = 0.02
delta = 1
gamma = 0.01
x0 = 20
y0 = 20

x0=20
y0=20
t0=0 
tf=30 

nt = int((tf-t0)/h) #número de pasos

def derivative(X, t, alpha, beta, delta, gamma):
    x, y = X
    dotx = alpha*x * (1 - x/K)-beta*x*y
    doty = -delta*y+gamma*x*y
    return np.array([dotx, doty])

Nt = nt
tmax = 30.
t = np.linspace(0.,tmax, Nt)
X0 = [x0, y0]
res = integrate.odeint(derivative, X0, t, args = (alpha, beta, delta, gamma))
x, y = res.T
(2, 1000)
plt.figure()
plt.grid()
plt.title("odeint method")
plt.plot(t, x, 'xb', label = 'Deer')
plt.plot(t, y, '+r', label = "Wolves")
plt.xlabel('Time t, [days]')
plt.ylabel('Population')
plt.legend()

plt.show()
```
